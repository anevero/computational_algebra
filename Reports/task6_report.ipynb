В задании требовалось написать программу, которая вычисляет максимальное по модулю собственное значение матрицы и соответствующий собственный вектор (возможно, пару максимальных по модулю значений и соответствующих векторов).

Решением данной задачи можно считать реализацию метода *GetPowerIterationResults* класса Matrix (см. соответствующий заголовочный файл).

Обработка связанных с этим заданием данных вынесена в отдельный класс, определенный в заголовочном файле *Task6*.

Реализованный алгоритм работает с **действительными** собственными значениями и соответствующими векторами. Если максимальное по модулю собственное значение является действительным, алгоритм вернет и его, и соответствующий собственный вектор. Если имеется два противоположных по знаку, но равных по модулю собственных значения, алгоритм вернет информацию об обоих из них (с соответствующими векторами). Для некоторых матриц алгоритм может вернуть информацию не только о максимальном, а еще и о каком-то другом собственном значении.

Работа с **комплексными** собственными значениями и соответствующими векторами не реализована.

## Описание алгоритма

Используется немного модифицированный алгоритм степенного метода. Используя псевдокод, каждую итерацию алгоритма можно описать следующим образом:

`
while ||Av - lambda*v|| > epsilon:
  v = Av
  u = Av
  lambda = +- sqrt(max(u))
  v = lambda * v + u
  v = v / max(v)
`

Значение $\lambda$ сходится к необходимому собственному значению, вектор $v$ сходится к соответствующему собственному вектору. Выбором знака $\lambda$ можно регулировать случаи, когда имеется два противоположных по знаку собственных значения.

Так как изначально непонятно, какова исходная матрица, и сколько собственных значений может найти алгоритм, для получения всех возможных результатов итерации с $\lambda$ и $-\lambda$ запускаются параллельно с ограничем на число итераций. Если алгоритм сошелся до того, как достиг ограничения, мы нашли собственное значение и собственный вектор. Возможны различные случаи, например:

* Не сошелся ни один из двух циклов. В этом случае мы можем констатировать, что либо матрица плохо подходит для использования этого метода (например, имеет два очень близких собственных значения), либо максимальное по модулю собственное значение является комплексным.
* Сошелся только один цикл. В этом случае мы получаем одно собственное значение и собственный вектор. При этом не гарантируется, что размерность подпространства векторов, соответствующих этому собственному значению, равна 1.
* Сошлись оба цикла. В этом случае обычно мы получим два противоположных по знаку собственных значения и соответствующие собственные вектора. Впрочем, для некоторых матриц мы можем получить и просто два разных собственных значения и вектора.

В перспективе используемый метод можно улучшить для нахождения комплексных собственных значений (запускать соответствующие итерации еще в одном потоке). Впрочем, насколько мне позволяет судить изученная литература, данный подход не очень распространен.
