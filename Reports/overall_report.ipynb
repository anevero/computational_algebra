# Отчет: лабораторная работа 1

*Неверо Андрей, 2 курс, 4 группа*

В данном файле приведены общие сведения о выполненных заданиях лабораторной работы. Более подробные описания конкретных заданий находятся в соответствующих файлах.

В качестве решения предоставлены файлы проекта, в котором реализованы необходимые классы и методы. Основой всех решений является класс *Matrix*, который задает множество различных операций над матрицами (начиная со сложения и умножения и заканчивая вычислением TLU-разложения, LDL-разложения и обратной матрицы).

Основу класса *Matrix* составляют методы, общие для любых матриц: операции сложения, умножения и подобные, перегруженный оператор вывода, методы доступа к приватным полям класса, метод для вычисления TLU-разложения (случайной квадратной матрицы), метод для решения системы линейных уравнений (заданной случайной квадратной матрицей и случайным вектором свободных членов), метод нахождения обратной матрицы (к случайной квадратной матрице).

Так как многие задания предполагают работу над какими-то специфическими типами матриц (например, почти треугольные или симметрические), в классе *Matrix* дополнительно определены методы работы именно с конкретными видами матриц (например, почти треугольными или симметрическими).

Большая часть кода сопровождается достаточно подробными комментариями о его назначении, логике. Присутствуют также асимптотические оценки производительности. Самые важные моменты продублированы в отчетах к соответствующим задачам.

Хотя в отчете приведены все необходимые данные касательно заданий, очевидно, что программу можно запустить и на каких-то других данных.

Хотя многие нестандартные ситуации обрабатываются с помощью механизма исключений (например, у вас не получится создать матрицу строк или найти обратную к вырожденной матрицу), программа отсекает далеко не все случаи передачи некорректных данных. Например, при использовании методов вычисления чего-либо в почти треугольной матрице, проверка на "почти треугольность" не проводится. Ответственность за переданные в функцию данные в этом случае лежит на пользователе.

## Структура проекта

Проект разделен на файлы, отвечающие за ту или иную часть функциональности.

* Папка *input_files*: содержит входные данные из условий задач.
* Папка *Reports*: содержит отчеты о всех выполненных заданиях (в том числе и этот общий отчет).
* Папка *Tasks*: содержит файлы с объявлениями и определениями классов, выполняющих задачи лабораторной. Например, класс для первой задачи содержит функцию расчета обратной матрицы к первой матрице, функцию расчета обратной матрице ко второй матрице, а также функцию, которая генерирует случайные даты и возвращает статистические данные о скорости и точности работы алгоритмов. Эти данные используются в соответствующем отчете. Фактически все функции во всех классах (кроме класса для пятой задачи) просто используют реализованные в классе *Matrix* методы для решения конкретной задачи. Следует отметить, что классы являются шаблонными: можно проверить, как будут вести себя алгоритмы в случае разных типов данных.
* Папка *Tests*: содержит файлы с функциями, которые использовались для тестирования работоспособности алгоритмов. Не все тесты обязательно будут пройдены, если вы их запустите (это ожидаемое поведение). Например, нахождение обратной матрицы к матрицам большого размера в любом случае даст сильную погрешность, и соответствующие тесты пройдены не будут.
* Папка *Utils*: содержит объявления и определения классов и функций, использовавшихся в качестве "утилит", например, функция, возвращающая случайные числа, или функции, которые в удобном формате выводят в файл данные, полученные при тестировании алгоритмов (эти данные затем использовались для построения графиков в отчетах).
* *CMakeLists.txt*: CMake-скрипт для сборки проекта. Собрать проект можно как вручную, так и с помощью какой-нибудь среды разработки. При создании проекта использовался [CLion](https://www.jetbrains.com/clion/), однако с CMake умеет работать и, например, [Visual Studio](https://docs.microsoft.com/en-us/cpp/build/cmake-projects-in-visual-studio?view=vs-2019). Чуть более подробные инструкции касательно Visual Studio расположены ниже.
* *main.cpp*: здесь можно вызывать любые функции любых классов.
* *Matrix.h*: заголовочный файл, содержащий объявления и определения методов шаблонного класса *Matrix*.

## Техническая часть

Стандарт C++ не гарантирует множество вещей, которые достаточно важны в этой лабораторной (например, он не устанавливает гарантированный размер типов вещественных чисел). В связи с этим различные компиляторы на практике могут выдавать разные результаты работы одного и того же алгоритма. Кроме того, очевидно, скорость работы программы часто напрямую зависит от мощности компьютера.

Некоторые характеристики аппаратного и программного обеспечения, на котором были получены предоставленные в отчетах результаты:
* Процессор: Intel Core i3-4000M (двухядерный процессор, выпущен в 2013 году)
* Оперативная память: 8 Гб
* ОС: Ubuntu Linux 18.04 x86_64
* Компилятор: clang++-10 / g++-9
* Уровень оптимизаций при компиляции: Release-конфигурация CMake с флагом -O3
* Испльзуемое соглашение представления вещественных чисел: IEEE 754
* Характеристики типа float: гарантированная точность при хранении - 6 значащих цифр, размер - 32 бита.
* Характеристики типа double: гарантированная точность при хранении - 15 значащих цифр, размер - 64 бита.
* Характеристики типа long double: гарантированная точность при хранении - 18 значащих цифр, размер - 80 бит (для выравнивания данных в памяти может быть увеличен до 128 бит без изменения точности числа).

## Важное замечание о работе алгоритмов под Windows

Все алгоритмы были написаны и тестировались под Linux. Хотя большинство функций должны предсказуемо работать и под Windows, корректный запуск и функционирование, например, многопоточных алгоритмов под Windows не гарантируется. Это связано с тем, что у каждого компилятора есть свои особенности реализации многопоточных алгоритмов, которые могут сильно влиять на их работоспособность и производительность. Например, реализация многопоточных алгоритмов стандартной библиотеки C++ в компиляторе MinGW для Windows достаточно специфична.

В случае сильной необходимости запуска программы под Windows можно воспользоваться [Windows Subsystem for Linux](https://docs.microsoft.com/en-us/windows/wsl/install-win10) и, например, [CLion](https://www.jetbrains.com/help/clion/how-to-use-wsl-development-environment-in-clion.html). Правда, производительность алгоритмов в этом случае наверняка упадет. Если нет необходимости работать с типом *long double*, можно попробовать скомпилировать проект с помощью компилятора Microsoft Visual Studio.

## Замечание о CMake и Visual Studio

Если вы решили скомпилировать проект с помощью Visual Studio, можно воспользоваться тем, что данная среда разработки умеет использовать CMake для компиляции проектов (хотя существуют и определенные особенности).

Для компиляции по умолчанию (debug-конфигурация без оптимизаций) достаточно просто открыть CMake-проект и запустить его. Открыть CMake-проект можно, например, так:
* В главном окне Visual Studio выбрать пункт "Указать путь к папке проекта" и указать путь к папке проекта (той, в которой лежит CMake-скрипт).
* В любом окне Visual Studio воспользоваться меню "Файл" и выбрать в нем соответствующий пункт.

Компиляция подобного проекта без оптимизаций уменьшит производительность решения в десятки раз, поэтому следует переключиться на release-конфигурацию и активировать оптимизации компилятора. Сделать это можно десятком различных способов. Например, можно открыть файл CMakeSettings.json, который Visual Studio автоматически сгенерирует на базе скрипта CMakeLists.txt, и добавить туда следующую конфигурацию:

`{
    "name": "x64-Release",
    "generator": "Ninja",
    "configurationType": "Release",
    "buildRoot": "${projectDir}\\out\\build\\${name}",
    "installRoot": "${projectDir}\\out\\install\\${name}",
    "cmakeCommandArgs": "-DCMAKE_CXX_FLAGS:STRING=\"/O2\"",
    "buildCommandArgs": "-v",
    "ctestCommandArgs": "",
    "inheritEnvironments": [ "msvc_x64_x64" ],
    "variables": []
}`

Обратите внимание на строку `cmakeCommandArgs`: в ней перечислены флаги компиляции (в примере - только флаг `/O2`). Если необходимо использовать и другие флаги, их следует добавить в эту строку. Например:

`"cmakeCommandArgs": "-DCMAKE_CXX_FLAGS:STRING=\"/O2 /Ob\"",`
