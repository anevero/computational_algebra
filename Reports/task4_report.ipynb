В задании требовалось написать программу, которая решает системы линейных уравнений с трехдиагональными матрицами.

Решением данной задачи можно считать реализацию метода *SolveSystem_Tridiagonal* класса Matrix (см. соответствующий заголовочный файл).

Обработка связанных с этим заданием данных вынесена в отдельный класс, определенный в заголовочном файле *Task4*.

## Основная идея алгоритма

Для экономии пространства трехдиагональные матрицы хранятся в виде матриц $n \times 3$. Первый элемент первой строки и последний элемент последней строки при подсчетах не учитываются и могут быть любыми.

Используется алгоритм правой прогонки. Сначала алгоритм проходится сверху вниз слева направо, зануляя элементы матрицы, лежащие слева от главной диагонали. Затем алгоритм проходится снизу вверх справа налево, зануляя элементы, лежащие справа от главной диагонали, и сразу считая соответствующую координату вектора решения.

Реализация выбора главного элемента для данного алгоритма видится неэффективной, так как это сломает структуру матрицы, благодаря которой мы можем вычислять решение за линейное время. В случае необходимости ее реализации можно попытаться оптимизировать алгоритм TLU-разложения для подобного вида матриц, однако в любом случае он будет работать дольше, чем рассмотренный выше алгоритм.
