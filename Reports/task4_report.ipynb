## Основная идея алгоритма

Для экономии пространства трехдиагональные матрицы хранятся в виде матриц $n \times 3$ (хранятся лишь три элемента каждой строки, начиная с элемента слева от главной диагонали). Первый элемент первой строки и последний элемент последней строки при подсчетах не учитываются и могут быть любыми.

Используется модифицированный алгоритм правой прогонки (с выбором главного элемента по столбцу). Матрица расширяется до размера $n \times 4$, последний столбец заполняется нулями. Все строки проходятся по очереди. На каждой итерации:

1. Элемент на главной диагонали текущей строки сравнивается с элементом слева от главной диагонали следующей строки (возможно, нулевым). Если он меньше, чем элемент на следующей строке, строки меняются местами. Если наибольший по модулю элемент из этих двух равен нулю, выбрасывается исключение. Заметим, что данные действия являются корректной реализацией выбора элемента по столбцу, так как во всех последующих строках все элементы в рассматриваемом столбце заведомо равны нулю.
2. Элемент слева от главной диагонали следующей строки зануляется путем отнимания текущей строки от следующей (возможно, домноженной на некоторый коэффициент). Нетрудно видеть, что четырех хранимых элементов каждой строки достаточно для проведения корректной операции отнимания (благодаря тому, что матрица является трехдиагональной).
3. Происходит переход к следующей итерации.

После этого мы получаем верхнетреугольную матрицу, которая в худшем случае имеет по три ненулевых элемента в каждой строке (элемент на главной диагонали и два элемента справа от нее). Для получения ответа достаточно реализовать обратный ход метода Гаусса, который, вследствие константного количества элементов в каждой строке, будет работать за линейное время.
