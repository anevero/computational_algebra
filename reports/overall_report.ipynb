# Отчет: лабораторные работы 1-2

*Неверо Андрей, 2 курс, 4 группа*

В данном файле приведены общие сведения о выполненных заданиях лабораторных работ. Более подробные описания конкретных заданий находятся в соответствующих файлах.

В качестве решения предоставлены файлы проекта, в котором реализованы необходимые классы и методы. Основой всех решений является класс *Matrix*, который задает множество различных операций над матрицами (начиная со сложения и умножения и заканчивая вычислением TLU, LDL, QR-разложений матрицы, алгоритмами по нахождению собственных значений и векторов).

Основу класса *Matrix* составляют методы, общие для любых матриц: операции сложения, умножения и подобные, перегруженный оператор вывода, методы доступа к приватным полям класса, метод для вычисления TLU-разложения (случайной квадратной матрицы), метод для решения системы линейных уравнений (заданной случайной квадратной матрицей и случайным вектором свободных членов), метод нахождения обратной матрицы (к случайной квадратной матрице).

Так как многие задания предполагают работу над какими-то специфическими типами матриц (например, почти треугольные или симметрические), в классе *Matrix* дополнительно определены методы работы именно с конкретными видами матриц (например, почти треугольными или симметрическими).

Вместе с классом *Matrix* доступна реализация класса *Polynomial*, для которого также определены как основные операции (сложение, умножение, дифференцирование), так и более сложные методы (нахождение корней с помощью метода бисекции и метода Ньютона).

Большая часть кода сопровождается достаточно подробными комментариями о его назначении, логике. Присутствуют также асимптотические оценки производительности. Самые важные моменты продублированы в отчетах к соответствующим задачам.

Хотя в отчете приведены все необходимые данные касательно заданий, очевидно, что программу можно запустить и на каких-то других данных.

Хотя многие нестандартные ситуации обрабатываются с помощью механизма исключений (например, у вас не получится создать матрицу строк или найти обратную к вырожденной матрицу), программа отсекает далеко не все случаи передачи некорректных данных. Например, при использовании методов вычисления чего-либо в почти треугольной матрице, проверка на "почти треугольность" не проводится. Ответственность за переданные в функцию данные в этом случае лежит на пользователе.

## Структура проекта

Проект разделен на файлы, отвечающие за ту или иную часть функциональности.

* Папка *input_files*: содержит входные данные из условий задач.
* Папка *reports*: содержит отчеты о всех выполненных заданиях (в том числе и этот общий отчет).
* Папка *tasks*: содержит файлы с объявлениями и определениями классов, выполняющих задачи лабораторной. Например, класс для первой задачи содержит функцию расчета обратной матрицы к первой матрице, функцию расчета обратной матрице ко второй матрице, а также функцию, которая генерирует случайные даты и возвращает статистические данные о скорости и точности работы алгоритмов. Эти данные используются в соответствующем отчете. Фактически все функции во всех классах просто используют реализованные в классе *Matrix* методы для решения конкретной задачи. Следует отметить, что классы являются шаблонными: можно проверить, как будут вести себя алгоритмы в случае разных типов данных.
* Папка *tests*: содержит файлы с функциями, которые использовались для тестирования работоспособности алгоритмов. Не все тесты обязательно будут пройдены, если вы их запустите (это ожидаемое поведение). Например, нахождение обратной матрицы к матрицам большого размера в любом случае даст сильную погрешность, и соответствующие тесты пройдены не будут.
* Папка *utils*: содержит объявления и определения классов и функций, использовавшихся в качестве "утилит", например, функция, возвращающая случайные числа, или функции, которые в удобном формате выводят в файл данные, полученные при тестировании алгоритмов (эти данные затем использовались для построения графиков в отчетах).
* *CMakeLists.txt*: CMake-скрипт для сборки проекта. Собрать проект можно как вручную, так и с помощью какой-нибудь среды разработки. При создании проекта использовался [CLion](https://www.jetbrains.com/clion/).
* *matrix.h*: заголовочный файл, содержащий объявления и определения методов шаблонного класса *Matrix*.
* *polynomial.h*: заголовочный файл, содержащий объявления и определения методов шаблонного класса *Polynomial*.

## Нумерация заданий 1-ой и 2-ой лабораторной

Так как решения обоих лабораторных работ фактически едины (реализации класса *Matrix* и *Polynomial*), объединена и нумерация заданий лабораторных.
Так, отчеты и классы, касающиеся первой лабораторной, нумеруются с 1 до 5, а отчеты и классы, касающиеся второй лабораторной - с 6 до 8. При этом шестой задачей считается первая задача второй лабораторной работы, и так далее.

## Техническая часть

Стандарт C++ не гарантирует множество вещей, которые достаточно важны в этой лабораторной (например, он не устанавливает гарантированный размер типов вещественных чисел). В связи с этим различные компиляторы на практике могут выдавать разные результаты работы одного и того же алгоритма. Кроме того, очевидно, скорость работы программы часто напрямую зависит от мощности компьютера.

Некоторые характеристики аппаратного и программного обеспечения, на котором были получены предоставленные в отчетах результаты:
* Процессор: Intel Core i3-4000M
* Оперативная память: 8 Гб
* ОС: Ubuntu Linux 18.04 x86_64
* Компилятор: clang++-10 / clang++-9
* Уровень оптимизаций при компиляции: Release-конфигурация CMake с флагом -O3
* Испльзуемое соглашение представления вещественных чисел: IEEE 754
* Характеристики типа float: гарантированная точность при хранении - 6 значащих цифр, размер - 32 бита.
* Характеристики типа double: гарантированная точность при хранении - 15 значащих цифр, размер - 64 бита.
* Характеристики типа long double: гарантированная точность при хранении - 18 значащих цифр, размер - 80 бит (для выравнивания данных в памяти может быть увеличен до 128 бит без изменения точности числа).

Отдельно отметим, что в проекте используются некоторые возможности C++20, которые на текущий момент поддерживаются далеко не во всех компиляторах. В связи с этим при сборке проекта настоятельно рекомендуется использовать упомянутые выше компиляторы.

## Важное замечание о работе алгоритмов под Windows

Все алгоритмы были написаны и тестировались под Linux. Хотя большинство функций должны предсказуемо работать и под Windows, корректный запуск и функционирование, например, многопоточных алгоритмов под Windows не гарантируется. Это связано с тем, что у каждого компилятора есть свои особенности реализации многопоточных алгоритмов, которые могут влиять на их работоспособность и производительность.

В случае сильной необходимости запуска программы под Windows можно воспользоваться [Windows Subsystem for Linux](https://docs.microsoft.com/en-us/windows/wsl/install-win10) и, например, [CLion](https://www.jetbrains.com/help/clion/how-to-use-wsl-development-environment-in-clion.html). Правда, производительность алгоритмов в этом случае наверняка упадет.
